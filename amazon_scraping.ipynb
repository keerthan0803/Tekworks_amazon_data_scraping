## step-1 Importing required libary

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.amazon.in/s?k=laptop&crid=KNEPYDTW3XX&sprefix=laptop%2Caps%2C601&ref=nb_sb_noss_2'
```

```python
header={'useragent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'}
```

```python
response=requests.get(url,headers=header)
response
```

```python
soup=BeautifulSoup(response.content,'html.parser')
soup
```

```python
products=soup.find_all('div',class_='sg-col-inner')
len(products)
```

```python
product_names=[]
product_prices=[]
for product in products:
    title_tag=product.h2
    if title_tag:
        product_names.append(title_tag.text.strip())
    price_tag=product.find("span", class_="a-price-whole")
    if price_tag:
        product_prices.append(price_tag.text.strip())
product_names
```

```python
product_prices=product_prices[:len(product_names)]
product_prices
```

```python
import re
```

```python
brand_list = ["Apple", "Asus", "Acer", "Dell", "HP", "Lenovo", "MSI", "Samsung", "Microsoft", "Razer", "Infinix", "Honor", "Realme", "Xiaomi"]
color_list = ["Black", "White", "Silver", "Gray", "Grey", "Blue", "Red", "Gold", "Green"]

def find_brand(title):
    if not title:
        return None
    for brand in brand_list:
        if title.lower().startswith(brand.lower()):
            return brand
    return title.split()[0]

def find_color(title):
    if not title:
        return None
    for color in color_list:
        if re.search(r"\b" + re.escape(color) + r"\b", title, re.IGNORECASE):
            return color.title()
    return None

def extract_ram(title):
    if not title:
        return None
    match = re.search(r"(\d+)\s*GB\s*RAM", title, re.IGNORECASE)
    if match:
        return match.group(1) + " GB"
    return None

def extract_storage(title):
    if not title:
        return None
    match = re.search(r"(\d+)\s*(TB|GB)\s*(SSD|NVMe|PCIe)", title, re.IGNORECASE)
    if match:
        size = match.group(1)
        unit = match.group(2).upper()
        return size + " " + unit + " SSD"
    match = re.search(r"(\d+)\s*(TB|GB)\s*SSD", title, re.IGNORECASE)
    if match:
        size = match.group(1)
        unit = match.group(2).upper()
        return size + " " + unit + " SSD"
    return None

def extract_windows(title):
    if not title:
        return None
    match = re.search(r"Windows\s*(\d+)", title, re.IGNORECASE)
    if match:
        return "Windows " + match.group(1)
    match = re.search(r"Win\s*(\d+)", title, re.IGNORECASE)
    if match:
        return "Windows " + match.group(1)
    return None

def extract_processor(title):
    if not title:
        return None
    patterns = [
        r"Intel\s+Core\s+i[3579][-\w]*",
        r"AMD\s+Ryzen\s+[3579]\s*\w*",
        r"Apple\s+M[123]\s*Max",
        r"Apple\s+M[123]",
        r"Intel\s+Pentium\s*\w*",
        r"Intel\s+Celeron\s*\w*",
    ]
    for pattern in patterns:
        match = re.search(pattern, title, re.IGNORECASE)
        if match:
            return match.group(0)
    return None

records = []
for product in products:
    title_tag = product.h2
    if not title_tag:
        continue
    title = title_tag.get_text(strip=True)
    price_tag = product.find("span", class_="a-price-whole")
    price = price_tag.get_text(strip=True) if price_tag else None
    rating_tag = product.find("span", class_="a-icon-alt")
    rating = None
    if rating_tag:
        rating_text = rating_tag.get_text(strip=True)
        rating_match = re.search(r"([\d.]+)\s*out of 5", rating_text, re.IGNORECASE)
        if rating_match:
            rating = float(rating_match.group(1))
    records.append({
        "product_title": title,
        "brand": find_brand(title),
        "ram": extract_ram(title),
        "storage_ssd": extract_storage(title),
        "windows": extract_windows(title),
        "color": find_color(title),
        "price": price,
        "rating": rating,
        "processor": extract_processor(title),
    })

laptop_data = pd.DataFrame(records)
laptop_data.head()
```
